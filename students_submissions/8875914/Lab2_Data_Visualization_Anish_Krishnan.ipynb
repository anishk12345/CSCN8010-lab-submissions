### Importing the required packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Using the TITANIC dataset in seaborn
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

### EDA

#### Scatterplot


```python
sns.scatterplot(data=titanic, x="age", y="fare")
```

#### Barplot

```python
sns.histplot(
    titanic,
    x="sex")
```

#### Pie Chart

```python
tmp = titanic['class'].value_counts().to_frame()
tmp = tmp.reset_index()
```

```python
colors = sns.color_palette('Set2')[0:3]
plt.pie(tmp['class'].to_list(), labels = tmp['index'].to_list(), colors = colors, autopct='%.0f%%')
plt.show()
```
